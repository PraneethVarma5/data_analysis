Drop unnamed column

```python
df.isnull().sum()#isnull() is used to return true if it has null value in each cell else of a column,false and sum() is used to count the no of true in a column
```

```python
df=df.drop("Unnamed: 0",axis=1)#drop() in pandas is used to remove specified labels from rows or columns in a DataFrame here axis=1 as i want to delete the column. if you want to delete row then axis=0
print(df.head())
```

change  weekly study hours column

```python
df["WklyStudyHours"] = df["WklyStudyHours"].str.replace("10-May","10-5")
df.head()
```

```python
plt.figure(figsize=(5,5))
ax=sns.countplot(data = df,x="Gender")
ax.bar_label(ax.containers[0])
plt.title("Gender Distributation")
plt.show()
```

```python
#from above chart we have analysed that the no of females in data is more than the no of males
```

```python
gb=df.groupby("ParentEduc").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
print(gb)
```

```python
plt.figure(figsize=(4,4))
sns.heatmap(gb,annot=True)
plt.title("Relationship b/w Parents education and student's score")
plt.show()
```

```python
#from the above chart we have concluded that the education  of the parents have a good impact on their scores
```

```python
gb1=df.groupby("ParentMaritalStatus").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
print(gb1)
```

```python
plt.figure(figsize=(4,4))
sns.heatmap(gb1,annot=True)
plt.title("Relationship b/w Parents Marital status and student's score")
plt.show()
```

```python
#from the above chart we have concluded that there is no/negligible impact on the student's score due to parents marital status
```

Distributation of Ethnic group

```python
groupA=df.loc[(df['EthnicGroup']=="group A")].count()
groupB=df.loc[(df['EthnicGroup']=="group B")].count()
groupC=df.loc[(df['EthnicGroup']=="group C")].count()
groupD=df.loc[(df['EthnicGroup']=="group D")].count()
groupE=df.loc[(df['EthnicGroup']=="group E")].count()

l=["group A","group B","group C","group D","group E"]
mlist=[groupA["EthnicGroup"],groupB["EthnicGroup"],groupC["EthnicGroup"],groupD["EthnicGroup"],groupE["EthnicGroup"]]

plt.pie(mlist,labels=l,autopct="%1.2f%%")
plt.title("Distribution of Ethnic groups")
plt.show()
```
